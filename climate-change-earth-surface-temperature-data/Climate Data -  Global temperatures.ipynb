```python
import requests
import json
from pandas.io.json import json_normalize
import time



def do_research(invalid_list_of_regions):
    regions_to_confirm = pd.DataFrame({'region_no_hit':[],'region_hit':[],'links':[],'schema_type':[], 'word_found':[],'similarityscore':[]})

    for region in invalid_list_of_regions:
        url = f"http://model.dbpedia-spotlight.org/en/annotate?text={region}"

        payload = {}
        headers= {
            'Accept':'application/json'
        }

        response = requests.request("GET", url, headers=headers, data = payload)
        print(response, region)
        if response.status_code == 200:
            result = json.loads(response.text.encode('utf8')) # get json data
            if 'Resources' in result:
                print(len(result['Resources']))
                test_data = json_normalize(result['Resources'])
                #print(test_data)
                test_data = test_data.rename(columns = {'@URI':'links','@types':'schema_type','@surfaceForm':'word_found','@similarityScore':'similarityscore'})

                for index, row in test_data.iterrows():
                    links = row['links']
                    schema = row['schema_type']
                    word = row['word_found']
                    score = row['similarityscore']
                    regions_to_confirm = regions_to_confirm.append({'region_hit':region,'links':links, 'schema_type':schema, 
                                                                            'word_found':word,'similarityscore':score}, ignore_index = True)
            time.sleep(5)   

        else:
            regions_to_confirm = regions_to_confirm.append({'region_no_hit':region}, ignore_index = True)
            time.sleep(5)
            
    return regions_to_confirm
        
        
```

```python
italy_regions_to_confirm = do_research(modified_italian_list)
```
