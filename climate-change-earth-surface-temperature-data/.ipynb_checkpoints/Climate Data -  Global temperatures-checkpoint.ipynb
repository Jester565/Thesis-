```python
# for loop over the regions 
# save the region name latitude and longitude

from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent = "geopy/1.22.0")
output = "CSV"

france_invalid_regions = []

for loc in confirm_france_list:
    location = geolocator.geocode(str(loc))
    
    try: 
        locations = [(location.latitude, location.longitude)]
        output_folder = r'C:\Users\rezan\Documents\Master in Applied Data Science\2nd Year\MASTER THESIS\GIT THESIS\geopy\new france' # if r'' the location of the script is where the files will be outputted.
        base_url = r"https://power.larc.nasa.gov/cgi-bin/v1/DataAccess.py?request=execute&identifier=SinglePoint&tempAverage=INTERANNUAL&parameters=T2M,RH2M&startDate=1981&endDate=2019&lat={latitude}&lon={longitude}&outputList={output}&userCommunity=SSE"

        for latitude, longitude in locations:
            api_request_url = base_url.format(longitude=longitude, latitude=latitude, output=output.upper())

            # Python Memory Object
            json_response = json.loads(requests.get(api_request_url).content.decode('utf-8'))

            # Selects the file URL from the JSON response
            csv_request_url = json_response['outputs'][output.lower()]
            #print('csv_request_url: ',csv_request_url)

            # Download File to Folder
            output_file_location = os.path.join(output_folder,str(loc)+'.csv') # os.path.basename(csv_request_url)+
            #print('output file location: ', output_file_location)
            urllib.request.urlretrieve(csv_request_url, output_file_location)
            
    except:
        france_invalid_regions.append(loc)
```

```python
import requests
import json
from pandas.io.json import json_normalize
import time



def do_research(invalid_list_of_regions):
    regions_to_confirm = pd.DataFrame({'region_no_hit':[],'region_hit':[],'links':[],'schema_type':[], 'word_found':[],'similarityscore':[]})

    for region in invalid_list_of_regions:
        url = f"http://model.dbpedia-spotlight.org/en/annotate?text={region}"

        payload = {}
        headers= {
            'Accept':'application/json'
        }

        response = requests.request("GET", url, headers=headers, data = payload)
        print(response, region)
        if response.status_code == 200:
            result = json.loads(response.text.encode('utf8')) # get json data
            if 'Resources' in result:
                print(len(result['Resources']))
                test_data = json_normalize(result['Resources'])
                #print(test_data)
                test_data = test_data.rename(columns = {'@URI':'links','@types':'schema_type','@surfaceForm':'word_found','@similarityScore':'similarityscore'})

                for index, row in test_data.iterrows():
                    links = row['links']
                    schema = row['schema_type']
                    word = row['word_found']
                    score = row['similarityscore']
                    regions_to_confirm = regions_to_confirm.append({'region_hit':region,'links':links, 'schema_type':schema, 
                                                                            'word_found':word,'similarityscore':score}, ignore_index = True)
            time.sleep(5)   

        else:
            regions_to_confirm = regions_to_confirm.append({'region_no_hit':region}, ignore_index = True)
            time.sleep(5)
            
    return regions_to_confirm
        
        
```
