```python
import pandas as pd
import numpy as np
import os
import re
```

# France

```python
rootdir = r"C:\Users\rezan\Documents\Master in Applied Data Science\2nd Year\MASTER THESIS\GIT THESIS\geopy\france"
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        filepath = os.path.join(subdir, file)
        with open(filepath) as f:
            lines = f.readlines()
        next_file_path = re.sub('france','france_cleaned',filepath)
        print(next_file_path)
        with open(next_file_path, "w") as outfile:
            outfile.write("".join(lines[11:]))
```
